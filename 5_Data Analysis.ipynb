## Leagues with more FIFA World Cup players

```python
df2.columns
```

```python
df2.head()
```

```python
leagues = df2[['year','club_origin']]
```

```python
leagues = leagues.rename(columns={'club_origin': 'league'})
```

```python
# Obtener una lista de años únicos en ambos DataFrames
unique_years = leagues['year'].unique()

# Crear una lista para almacenar los resultados
results = []

# Iterar a través de cada año único
for year in unique_years:
    # Obtener los valores únicos de club_origin para el año actual
    club_origin_values = leagues[leagues['year'] == year]['league'].unique()
    
    # Verificar si hay al menos un valor que coincide en national_team
    for club in club_origin_values:
        if club in nat_teams[nat_teams['year'] == year]['national_team'].values:
            results.append({'year': year, 'league': club, 'country_played_wc': True})
        else:
            results.append({'year': year, 'league': club, 'country_played_wc': False})

# Crear un DataFrame a partir de la lista de resultados
results_df = pd.DataFrame(results)



# Imprimir el DataFrame de resultados
print(leagues)
# Este código verificará para cada año si cada valor único en la columna "club_origin" del primer 
# DataFrame coincide con al menos un valor en la columna "national_team" del segundo DataFrame y 
# almacenará los resultados en un nuevo DataFrame llamado "results_df". Los resultados se marcan como True
# si hay coincidencia y como False si no hay coincidencia.
```

```python
# Merge to add results as a new column in 'Leagues'
leagues = leagues.merge(results_df, on=['year', 'league'], how='left')
```

```python
# Load the shape file 
world = gpd.read_file('world_new\\world.shp')

# Crear un GeoDataFrame uniendo tus datos y la geometría
gdf = world.merge(leagues2022, left_on='country', right_on ='league', how = 'left')

# Fill Nan values 
gdf['year'].fillna(2022.0, inplace=True) 
gdf['league'].fillna(gdf['country'], inplace=True)
gdf['country_played_wc'].fillna(False, inplace=True)
gdf['players'].fillna(0.0, inplace=True)

# Create a choroplets map
fig, ax = plt.subplots(figsize=(16, 8))

# Cambia el color de fondo del área exterior
ax.set_facecolor('lightgray')

gdf.plot(column='players', cmap='YlGnBu', legend=True, ax=ax, edgecolor='black', linewidth=0.2)
plt.title('Leagues with more World Cup players 2022')
plt.show()
```

```python
import matplotlib.colors as mcolors

# Define tus valores de escala personalizada
custom_scale = [0, 10, 20, 50, 80, 110, 140,170]  # Por ejemplo, puedes establecer estos valores como tu escala personalizada

# Crea una paleta de colores personalizada basada en tu escala
custom_colors = plt.cm.YlGnBu(np.linspace(0, 1, len(custom_scale)))

# Crea la paleta de colores personalizada
cmap_custom = mcolors.ListedColormap(custom_colors)

# Define los límites de la paleta de colores personalizada
norm_custom = plt.Normalize(vmin=custom_scale[0], vmax=custom_scale[-1])

# Luego, utiliza esta paleta de colores personalizada en tu mapa de coropletas
fig, ax = plt.subplots(figsize=(16, 8))

# Cambia el color de fondo del área exterior
ax.set_facecolor('lightgray')

gdf.plot(column='players', cmap=cmap_custom, norm=norm_custom, legend=True, ax=ax, edgecolor='black', linewidth=0.2)


plt.title('Leagues with more World Cup players 2022')
plt.show()
```
